### <mark> Observation </mark>

#### Our dataset contains 16 columns and 1,465 rows.
#### All 16 columns are of type object.
#### We expect the following columns to be numeric:
#### discounted_price, actual_price, discount_percentage, rating, and rating_count.
#### Therefore, the next step is to convert these columns to numeric types for accurate analysis.

```python
df[['discounted_price','actual_price']].head()
```

```python
for col in ['discounted_price', 'actual_price']:
    df[col] = df[col].str.replace('‚Çπ', '', regex=False)
    df[col] = df[col].str.replace(',', '', regex=False)
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df[['discounted_price','actual_price']].head()
```

```python
df[['rating','rating_count']].head()
```

```python
for col in ['rating', 'rating_count']:
    df[col] = df[col].replace(',', '', regex=False)
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df[['rating','rating_count']].head()
```

```python
df['discount_percentage'].head()
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '', regex=False)
df['discount_percentage']=pd.to_numeric(df['discount_percentage'],errors='coerce')
```

# Uncovering Trends :

```python
top_categories = df['category'].value_counts().head(15)

plt.figure(figsize=(14, 6))
sns.barplot(x=top_categories.values, y=top_categories.index, palette="crest")

plt.title("Top 20 Product Categories")
plt.xlabel("Count")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

### Outcome of the above visual:

The category plot indicates that the Electronics segment holds the strongest market presence.
Within this segment, the top-performing sub-categories are:

Computers & Accessories ‚Üí Accessories & Peripherals ‚Üí Cables & Accessories ‚Üí Cables ‚Üí USB Cables

Among these core categories, USB Cables show the highest dominance in performance.

Additionally, after Smartwatches, the Smartphone and Mixer Grinder categories also show strong performance trends.

```python
plt.title('Discounted Percentage vs Actual price Distribution ')
sns.scatterplot(data=df,x='discount_percentage',y = 'actual_price')
```

## Outcome 

The plot indicates that a maximum discounts are given to products that usually lie between the actual_price range of 0 to 10,000 INR . It also indicates that the products are highly accumulated in price range under 10,000 INR.

```python
plt.title("Rating vs Discounted Percentage Distribution ")
sns.scatterplot(data=df,x='rating',y = 'discount_percentage')
```

## Outcome 

There is no strong relationship between rating and discount percentage.
Products with high discounts do not necessarily have high ratings, and products with low discounts can still have good ratings. Most ratings are clustered between 3.5 to 4.5, regardless of how big the discount is.

i.e Stores give discounts mainly to increase sales or clear inventory, not because the product quality (rating) is high or low.

```python
plt.title("Correlation Between Numeric Features")
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

## Outcome

There is a very strong positive relationship between actual price and discounted price (0.96), indicating that discounts are applied proportionally based on original price.

Discount percentage has a slight negative correlation with discounted price (-0.24), meaning lower-priced products often show higher discount percentages.

Product ratings and rating counts show very weak correlations with price and discounts, suggesting that customer satisfaction and engagement do not depend on pricing strategies.
