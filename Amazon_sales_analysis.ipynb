### <mark> Observation </mark>

#### Our dataset contains 16 columns and 1,465 rows.
#### All 16 columns are of type object.
#### We expect the following columns to be numeric:
#### discounted_price, actual_price, discount_percentage, rating, and rating_count.
#### Therefore, the next step is to convert these columns to numeric types for accurate analysis.

```python
df[['discounted_price','actual_price']].head()
```

```python
for col in ['discounted_price', 'actual_price']:
    df[col] = df[col].str.replace('₹', '', regex=False)
    df[col] = df[col].str.replace(',', '', regex=False)
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df[['discounted_price','actual_price']].head()
```

```python
df[['rating','rating_count']].head()
```

```python
for col in ['rating', 'rating_count']:
   # df[col] = df[col].str.replace('₹', '', regex=False)
    df[col] = df[col].str.replace(',', '', regex=False)
    df[col] = pd.to_numeric(df[col], errors='coerce')
    
```

```python
df[['rating','rating_count']].head()
```

```python
df['discount_percentage'].head()
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '', regex=False)
df['discount_percentage']=pd.to_numeric(df['discount_percentage'],errors='coerce')
```

```python
df['discount_percentage'].head()
```

```python
df.info()
```

### <mark> Observation </mark>

#### We have successfully converted the following columns to numeric types:
#### discounted_price, actual_price, rating, rating_count, and discount_percentage_numeric
